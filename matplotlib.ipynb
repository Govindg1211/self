## saving a chart

```python
import matplotlib.pyplot as plt

x = ["part1","part2","part3","part4","part5"]
y = [55,45,78,32,62]

plt.bar(x,y)
plt.xlabel("parts of the data")
plt.ylabel("popularity")
plt.title("distribution of popularity")
plt.savefig("bar.png") # you have to specify the extension of the image  ("name.extension")
plt.show()
```
